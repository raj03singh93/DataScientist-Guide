```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.NaT,
                            pd.NaT]})
```

```python
df
```

```python
# drop null column/row 
'''
NOTE: How mean at what condition
'''
df.dropna(how = "all", axis = 1)
```

```python
df
```

```python
# Drop duplicate row
df.drop_duplicates(subset = ["name"])
```

```python
df.drop_duplicates(subset = ["name"], keep = "last")
```

```python
# shape of dataframe
df.shape
```

```python
# Get count of unique values
df["toy"].value_counts()
```

```python
# Get count of unique values(remove duplicates)
df["toy"].value_counts(dropna = False)
```

```python
# checks column which are not null
df.notnull()
```

```python
df.isnull()
```
